```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

print('\nload dataset\n')
df = pd.read_csv('netflix_titles.csv')
print(df.head())

print('\nBasic Cleaning')
print('\nchecking missing values\n')
print(df.isna().sum())

print('\nDrop missing rows from director or cast if needed\n')
print(df.dropna(subset=['director','cast'], inplace=True))

print('\nseaborn visualizes\n')
print('\nBar Plot â†’ Count of Movies vs TV Shows\n')
sns.countplot(data=df, x='type')
plt.title("Count of Movies vs TV Shows")
plt.xlabel("Type")
plt.ylabel("Count")
plt.show()

print('\nBox Plot â†’ Movie durations (just for Movie type)\n')
movies = df[df['type'] == 'Movie'].copy()
movies['duration'] = movies['duration'].str.replace('min','').astype(float)
sns.boxplot(data=movies, y='duration')
plt.title("Duration of Movies")
plt.ylabel("Minutes")
plt.show()

print('\nScatter Plot â†’ Release year vs duration\n')
sns.scatterplot(data=movies, x='release_year', y='duration')
plt.title("Release Year vs Movie Duration")
plt.xlabel("Year")
plt.ylabel("Duration (min)")
plt.show()

print('\nHeatMap->Correlation matrix (only numeric field)\n')
sns.heatmap(movies[['duration']].corr(), annot=True)
plt.title("Correlation Heatmap")
plt.show()

print('\nðŸ–Š Matplotlib Custom Labels\n')
plt.figure(figsize=(8,6))
plt.hist(movies['duration'], bins=20, color='orange')
plt.title("Movie Duration Distribution")
plt.xlabel("Duration (minutes)")
plt.ylabel("Number of Movies")
plt.grid(True)
plt.show()
```
